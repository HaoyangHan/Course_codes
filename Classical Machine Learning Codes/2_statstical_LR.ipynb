```python
# mathmatical calculation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import math
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("student_debt.csv", header=None)
X = data[0]
y = data[1]
m = len(y)
plt.scatter(X,y)
plt.title("Picture 1: scatter plot of dataset: x, y")
plt.show()
```

```python
X_mean = np.mean(X)
y_mean = np.mean(y)
        
def Sxx(X, y):
    X_ = []
    for i in range(len(X)):
        X_.append((X[i]-X_mean)**2)
    return sum(X_)

def Sxy(X, y):
    Xy_ = []
    for i in range(len(X)):
        Xy_.append((X[i]-X_mean)* (y[i]-y_mean))
    return sum(Xy_)

def beta_1(X, y):
    return Sxy(X, y)/Sxx(X, y)

def beta_0(X, y):
    return y_mean - beta_1(X, y) * X_mean
```

```python
beta0 = beta_0(X,y)
beta1 = beta_1(X,y)

x1 = np.linspace(2004,2014,100)
y1 = beta0 + beta1 * x1


prediction = beta0 + beta1 * 2050
print("the prediction value of year 2050 is:", prediction)
plt.scatter(X,y)
plt.plot(x1, y1, color = 'red')
plt.title("Linear Regression result")
plt.show()


```
