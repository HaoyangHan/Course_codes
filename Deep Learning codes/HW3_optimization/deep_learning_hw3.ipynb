````python
# YOUR CODE GOES HERE.
demo1 = superlearn_setup.Setup(x,y)
demo1.choose_features(name = 'multilayer_perceptron',layer_sizes = [784,30,20,10,10],activation = 'relu')

# YOUR CODE GOES HERE.
# choose normalizer
demo1.choose_normalizer(name = 'standard')

# choose cost
demo1.choose_cost(name = 'multiclass_softmax')
````
